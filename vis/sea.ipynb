```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('flights')
sns.lineplot(x='year', y='passengers',data=df)
plt.show()
```

```python
sns.barplot(data=sns.load_dataset('tips'),x='sex', y='total_bill')
plt.show()
```

```python
sns.countplot(x='day', data=sns.load_dataset('tips'))
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=sns.load_dataset('tips'))
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=sns.load_dataset('tips'))
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=sns.load_dataset('tips'), jitter=True)
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data=sns.load_dataset('tips'))
plt.title('Swarmplot of Total Bill by Day')
plt.show()
```

```python
flights = df.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

```python
sns.jointplot(data=sns.load_dataset('tips'), x='total_bill', y='tip')
plt.show()
```
