```python
import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [2, 4, 6]
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

```python
plt.bar(['A', 'B', 'C'], [10, 20, 15])
plt.title("Bar Plot")
plt.show()
```

```python
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
data = [np.random.normal(size=100) for _ in range(4)]
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

```python
labels = 'A', 'B', 'C'
sizes = [15, 30, 55]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
data = np.random.rand(5, 5)
plt.imshow(data, cmap='viridis', interpolation='none')
plt.colorbar()
plt.title("Heatmap")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2)
axs[0].plot([1, 2], [3, 4])
axs[1].bar([1, 2], [4, 5])
plt.suptitle("Subplots")
plt.show()
```
